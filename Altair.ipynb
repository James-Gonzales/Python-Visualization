```python
import altair as alt
import pandas as pd
from vega_datasets import data
```

```python
df=pd.DataFrame({
    'city': ['Seattle', 'Seattle', 'Seattle', 'New York','New York','New York','Chicago','Chicago','Chicago'],
    'month': ['Apr', 'Aug', 'Dec', 'Apr', 'Aug', 'Dec', 'Apr', 'Aug', 'Dec'],
    'precip': [2.68, 0.87, 5.31, 3.94, 4.13, 3.58, 3.62, 3.98, 2.56]
})
df
```

```python
alt.Chart(df).mark_point()
```

```python
alt.Chart(df).mark_point().encode(x='precip',y='city')
```

```python
alt.Chart(df).mark_point().encode(x='average(precip)',y='city')
```

```python
alt.Chart(df).mark_point(color="firebrick").encode(alt.X('precip', scale=alt.Scale(type='log'), axis=alt.Axis(title='Log-Scaled Values')),
alt.Y('city',axis=alt.Axis(title='Category')),
)
```

```python
cars=data.cars()
cars
```

```python
alt.Chart(cars).mark_bar().encode(
    alt.X("Miles_per_Gallon:Q", bin=True),
    y='count()',
)
```

```python
base = alt.Chart(cars)
 
bar = base.mark_bar().encode(
    x=alt.X('Miles_per_Gallon:Q', bin=True, axis=None),
    y='count()'
)
rule = base.mark_rule(color='red').encode(
    x='mean(Miles_per_Gallon):Q',
    size=alt.value(5)
)
bar+rule
```

```python
alt.Chart(cars).mark_bar().encode(
 alt.X("Horsepower:Q", bin=True),
 y='count()',
 row='Origin'
)
```

```python
alt.Chart(cars).transform_fold(
        ['Displacement', 'Horsepower','Cylinders']
).mark_area(
    opacity=0.3,
    interpolate='step'
).encode(
    alt.X('Miles_per_Gallon:Q', bin=alt.Bin(maxbins=100)),
    alt.Y('count()', stack=None),
    alt.Color('Displacement:N')
)
```

```python
source = data.wheat()   
source
```

```python
alt.Chart(source).mark_bar().encode(
    x='year:O',
    y="wheat:Q",
 
        color=alt.condition(
        alt.datum.year == 1810,  # If the year is 1810 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(width=600)
```

```python
bars = alt.Chart(source).mark_bar().encode(
    x='wheat:Q',
    y="year:O"
)
 
text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text='wheat:Q'
)
(bars + text).properties(height=900)
```

```python
bar = alt.Chart(source).mark_bar().encode(
    x='wages',
    y='wheat'
).properties(
    width=alt.Step(40)  # controls width of bar.
)
tick = alt.Chart(source).mark_tick(
    color='red',
    thickness=2,
    size=40 * 0.9,  # controls width of tick.
).encode(
    x='wages',
    y='wheat'
)
bar + tick
```

```python
   alt.Chart(source).mark_bar().encode(
    x='sum(wages):Q',
    y=alt.Y('year:N', sort='-x')
)
```

```python
alt.Chart(source).mark_bar().encode(
    x='year',
    y='sum(wages)',
    color='wheat'
)
```

```python
bars = alt.Chart(source).mark_bar().encode(
    x=alt.X('sum(wages):Q', stack='zero'),
    y=alt.Y('wheat:N'),
    color=alt.Color('year')
)
 
text = alt.Chart(source).mark_text(dx=-15, dy=3, color='white').encode(
    x=alt.X('sum(wages):Q', stack='zero'),
    y=alt.Y('wheat:N'),
    detail='year:N',
    text=alt.Text('sum(wages):Q', format='.1f')
)
 
bars + text
```
