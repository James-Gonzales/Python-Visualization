```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
rs = sns.load_dataset("iris")
rs
```

```python
brplot = sns.barplot(x = 'species', y = 'petal_width',data = rs,palette = "Set1")
brplot.set(xlabel = 'Species', ylabel = 'Petal Width' )
```

```python
sns.boxplot(x = 'species', y = 'petal_width',data = rs,palette = "coolwarm")
```

```python
sns.boxplot(x = 'species', y = 'petal_width',data = rs,hue ='species',  palette = "Set1")
```

```python
sns.violinplot(x = 'species', y = 'petal_width',data = rs,palette = "Set1")
```

```python
sns.scatterplot(data = rs , x = "sepal_length", y = "petal_width")
```

```python
sns.catplot(x="species", y="petal_width",  data = rs)
```

```python
sns.histplot(rs['petal_width'], color = 'b',kde = False)
```

```python
sns.histplot(rs['petal_width'], color = 'b',kde = True)
```

```python
sns.distplot(rs['sepal_length'],hist=False, bins=10)
```

```python
sns.distplot(rs['petal_width'],  kde = True, bins = 10)
```

```python
sns.stripplot(x='species', y="petal_width", data=rs)
```

```python
sns.set(style = 'whitegrid') 
sns.stripplot(x="sepal_length", y="petal_length", hue="species", data=rs)
```

```python
sns.swarmplot(x='species', y="petal_width", data=rs)
```

```python
sns.swarmplot(x="sepal_length", y="sepal_width", hue="species", data=rs)
```

```python
sns.regplot(x="petal_length", y="sepal_length", data=rs)
```

```python
sns.lmplot(x="petal_length", y="petal_width", data=rs)
```

```python
sns.lmplot(x="petal_length", y="petal_width", data=rs, x_estimator=np.mean)
```

```python
rs.corr()
```

```python
sns.heatmap(rs.corr(), cmap="coolwarm", annot=True)
```

```python
sns.lineplot(x="petal_length", y="petal_width",data = rs)
```

```python
rs.plot(figsize = (12,5))
plt.show()
```

```python
rs.plot(figsize = (12,5), linewidth = 5,fontsize = 10)
plt.xlabel("species", fontsize = 10)
```

```python
sns.pairplot(rs)
```
